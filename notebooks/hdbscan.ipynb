## HDBSCAN + SBD

```python
import numpy as np
import matplotlib.pyplot as plt
from collections import defaultdict
```

```python
import hdbscan
import scipy.sparse
import scipy.stats
import sklearn.cluster
from sklearn.metrics.pairwise import pairwise_distances
from sklearn.neighbors import NearestNeighbors
```

```python
import sys
sys.path.append('../')
from tsdr.testseries import sockshop
from tsdr.clustering.sbd import sbd 
```

```python
cases = sockshop.testcases_of_sockshop
ts = np.array([cases[i]['datapoints'] for i in range(len(cases))])

distance_matrix = pairwise_distances(ts, metric=sbd)
print(distance_matrix)

clusterer = hdbscan.HDBSCAN(
    min_cluster_size=2,
    metric='precomputed',
    allow_single_cluster=True,
).fit(distance_matrix)
labels = clusterer.labels_

label_to_cases = defaultdict(list)
for i, case in enumerate(cases):
    label_to_cases[labels[i]].append(case)
display(label_to_cases.keys())

fig, axs = plt.subplots(len(label_to_cases.keys()), ncols=2)
fig.set_size_inches(18, 1*len(cases))

for cases, ax in zip(label_to_cases.values(), axs.flatten()):
    ax.grid()
    for case in cases:
        ax.plot(case['datapoints'], label=case['name'])
    ax.legend()
```
