```python
import sys
sys.path.append('../')
%reload_ext autoreload
%autoreload 2

import numpy as np
import pandas as pd
from scipy.spatial.distance import hamming

from tsdr import tsdr 
from tsdr.clustering.sbd import sbd

issued_metrics_df = pd.DataFrame()
for k, vals in issued_metrics.items():
    res = tsdr.ar_based_ad_model(np.array(vals))
    issued_metrics_df[k] = res.binary_scores()

with pd.option_context('display.max_rows', None, 'display.max_columns', None):
    display(issued_metrics_df)

cinfo, removes = tsdr.hierarchical_clustering(
    issued_metrics_df,
    hamming,
    0.2 / 120,
    'medoid',
)
print(cinfo)
```

```python
filepath = '/datasets/argowf-chaos-hg68n/2021-12-09-argowf-chaos-hg68n-carts_pod-memory-hog_0.json'
data_df, _, metrics_meta = tsdr.read_metrics_json(filepath)
```

```python
param = {
    'tsifter_step1_ar_regression': 'n',
    'tsifter_step1_cv_threshold': 0.05,
    'tsifter_step2_clustering_threshold': 0.5,
    'tsifter_step2_clustered_series_type': 'binary_anomaly_score',
    'tsifter_step2_clustering_dist_type': 'sbd',
    'tsifter_step2_clustering_choice_method': 'maxsum',
    'tsifter_step2_clustering_linkage_method': 'single',
}
reducer = tsdr.Tsdr(tsdr.ar_based_ad_model, **param)
elapsed_time_by_step, reduced_df_by_step, metrics_dimension, clustering_info = reducer.run(
    series=data_df,
    max_workers=4,
)
```
